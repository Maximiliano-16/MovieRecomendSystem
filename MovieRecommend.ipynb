Мы можем использовать средний рейтинг фильма в качестве оценки, но это будет не совсем справедливо, так как фильм со средним рейтингом 8,9 и всего 3 голосами не может считаться лучше, чем фильм со средним рейтингом 7,8, но 40 голосами. Поэтому я буду использовать взвешенный рейтинг IMDB (wr), который выглядит следующим образом :
Например, формула $$ Wr = (v/(v+m)*R) + (m / (v+m) * C) , $$ где \\( a=0 \\).

где,
\\( v \\) - количество голосов за фильм;  
\\( m \\) - минимальное количество голосов, необходимое для попадания в таблицу;  
\\( R \\) - средний рейтинг фильма;  
\\( C \\) - средняя оценка по всему отчету.  
У нас уже есть v(vote_count) и R (vote_average), а C можно рассчитать как


```python
C = movies_df_new['vote_average'].mean()
print(C)
```

Итак, средняя оценка всех фильмов составляет около 5.6 по 10-балльной шкале. Следующий шаг - определить подходящее значение m, минимальное количество голосов, необходимое для включения в таблицу. Мы будем использовать 50-й процентиль в качестве отсечки. Другими словами, чтобы фильм попал в чарт, он должен набрать больше голосов, чем хотя бы половина фильмов в списке.

```python
movies_df_new['vote_count']
```

```python
m = movies_df_new['vote_count'].quantile(0.7)
print(m)
```

Теперь мы можем отфильтровать фильмы, которые попадают в финальную таблицу

```python
q_movies = movies_df_new.copy().loc[movies_df_new['vote_count'] >= m]
q_movies.shape
```

Получилось 13837 фильма подходят для дальнейших шагов. Теперь нам нужно рассчитать нашу метрику для каждого фильма, прошедшего отбор. Для этого мы определим функцию weighted_rating() и определим новую характеристику final_score, значение которой мы вычислим, применив эту функцию к нашему DataFrame квалифицированных фильмов:

```python
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    # Calculation based on the IMDB formula
    return (v/(v+m) * R) + (m/(m+v) * C)
```

```python
q_movies['score'] = q_movies.apply(weighted_rating, axis=1)
```

```python
#Sort movies based on score calculated above
q_movies = q_movies.sort_values('score', ascending=False)

#Print the top 15 movies
q_movies[['title', 'vote_count', 'vote_average', 'score']].head(10)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words='english')

tfidf_matrix = tfidf.fit_transform(q_movies['overview'])

tfidf_matrix.shape
```

```python
# Import linear_kernel
from sklearn.metrics.pairwise import linear_kernel

# Расчитаем косинусное сходство между фильмами
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

Определим функцию, которая принимет на вход название фильма и выдает список из 10 наиболее похожих фильмов. 

```python
#обратная связть индексов и названий фильмов
indices = pd.Series(q_movies.index, index=q_movies['title']).drop_duplicates().reset_index()
```

```python
indices
```

```python
def get_recommendations(title, cosine_sim=cosine_sim):
    try:
        # Проверим, есть ли фильм в списке
        if title not in indices['title'].values:
            return f"Фильм с названием '{title}' не найден в базе данных."

        # Получаем индекс фильма по названию
        idx = indices[indices['title'] == title].index[0]

        # Вычисляем оценки сходства с другими фильмами
        sim_scores = list(enumerate(cosine_sim[idx]))

        # Сортируем фильмы по убыванию сходства
        sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

        # Берем топ-10 фильмов (исключая сам фильм)
        sim_scores = sim_scores[1:11]

        # Получаем индексы рекомендованных фильмов
        movie_indices = [i[0] for i in sim_scores]

        # Возвращаем названия рекомендованных фильмов
        return q_movies['title'].iloc[movie_indices]

    except IndexError:
        return "Не удалось найти фильм с таким индексом, проверьте корректность данных."
    except KeyError:
        return "Произошла ошибка с ключами данных, возможно, данные имеют неверный формат."
    except Exception as e:
        return f"Произошла неожиданная ошибка: {str(e)}"
```
